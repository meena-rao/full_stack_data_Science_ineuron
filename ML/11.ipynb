### 1. Given X be a discrete random variable with the following PMF
#### p(x) = 0.1 for x=0.2 and  0.2 for x=0.4 and 0.2 for x =0.5, and 0.3 for x=0.8 and 0.2 for x=1 , 0 otherwise. 

#### 1. Find the range RX of the random variable X. 

#### 2. Find P(X ≤ 0.5)  
#### 3. Find P(0.25<X<0.75) 
#### 4. P(X = 0.2|X<0.6)


The range of the random variable X is the set of all possible values that X can take. In this case, X can take the values 0.2, 0.4, 0.5, 0.8, and 1. Therefore, the range of X is RX = {0.2, 0.4, 0.5, 0.8, 1}.

To find P(X ≤ 0.5), we need to sum the probabilities of all values of X that are less than or equal to 0.5. In this case, we have:

P(X ≤ 0.5) = P(X = 0.2) + P(X = 0.4) + P(X = 0.5) = 0.1 + 0.2 + 0.2 = 0.5

Therefore, the probability of X being less than or equal to 0.5 is 0.5.

To find P(0.25<X<0.75), we need to sum the probabilities of all values of X that are between 0.25 and 0.75, exclusive. In this case, we have:
P(0.25<X<0.75) = P(X = 0.4) + P(X = 0.5) = 0.2 + 0.2 = 0.4

Therefore, the probability of X being between 0.25 and 0.75, exclusive, is 0.4.

To find P(X = 0.2|X<0.6), we need to find the conditional probability of X being equal to 0.2, given that X is less than 0.6. In other words, we need to find the probability of X being equal to 0.2, given that X can only take the values 0.2, 0.4, and 0.5. In this case, we have:
P(X = 0.2|X<0.6) = P(X = 0.2 and X<0.6) / P(X<0.6)

We can find P(X<0.6) by summing the probabilities of all values of X that are less than 0.6:

P(X<0.6) = P(X = 0.2) + P(X = 0.4) + P(X = 0.5) = 0.1 + 0.2 + 0.2 = 0.5

To find P(X = 0.2 and X<0.6), we can simply use the probability of X being equal to 0.2, since X<0.6 is a subset of X being equal to 0.2:

P(X = 0.2 and X<0.6) = P(X = 0.2) = 0.1

Therefore, we have:

P(X = 0.2|X<0.6) = P(X = 0.2 and X<0.6) / P(X<0.6) = 0.1 / 0.5 = 0.2

Therefore, the conditional probability of X being equal to 0.2, given that X is less than 0.6, is 0.2.

### 3. In an exam, there were 20 multiple-choice questions. Each question had 44 possible options. The probability of this happening is (10 choose X-10) * (1/44)^(X-10) * (43/44)^(10-(X-10)), since he needs to choose X-10 questions out of the 10 he knew the answers to correctly, and then choose the correct option for each of those questions and the incorrect option for the remaining questions.

For the remaining 10 questions, the probability of getting all of them wrong is (33/34)^10, and the probability of getting k of them correct is (10 choose k) * (1/34)^k * (33/34)^(10-k).

Therefore, we have:

p(X) = (10 choose X-10) * (1/44)^(X-10) * (43/44)^(10-(X-10)) * [sum from k=0 to 10 of (10 choose k) * (1/34)^k * (33/34)^(10-k)]

Now, we can find P(X>15) by summing the probabilities for X=16, 17, 18, 19, and 20:

### 4. The number of students arriving at a college between a time interval is a Poisson random variable. On average, 10 students arrive per hour. Let Y be the number of students arriving from 10 am to 11:30 am. What is P(10<Y≤15)?

The number of students arriving in a given time interval is a Poisson random variable with parameter λ, where λ is the average number of arrivals per unit time. In this case, λ = 10 since there are 10 arrivals per hour on average.

Let's define Y as the number of students arriving between 10 am and 11:30 am. Since this time interval is 1.5 hours long, the number of arrivals during this interval is also a Poisson random variable, but with parameter λ' = 1.5*10 = 15, since we expect 15 arrivals on average during this time interval.

Now, we need to find P(10<Y≤15), which is the probability that the number of arrivals during the time interval is greater than 10 but less than or equal to 15. This can be written as:

P(10<Y≤15) = P(Y≤15) - P(Y≤10)

Using the Poisson distribution, we can compute these probabilities as follows:

P(Y≤15) = ∑(k=0 to 15) (e^-λ' * λ'^k) / k!

P(Y≤10) = ∑(k=0 to 10) (e^-λ' * λ'^k) / k!

Using a calculator, we find:

P(Y≤15) ≈ 0.983
P(Y≤10) ≈ 0.275

Therefore,

P(10<Y≤15) ≈ P(Y≤15) - P(Y≤10) ≈ 0.983 - 0.275 ≈ 0.708

So, the probability that the number of students arriving between 10 am and 11:30 am is greater than 10 but less than or equal to 15 is approximately 0.708.

### 5.Two independent random variables, X and Y,are given such that X~Poisson(α) and Y~Poisson(β). State a new random variable as Z = X + Y. Find out the PMF of Z.
Since X and Y are independent Poisson random variables with parameters α and β, respectively, the PMFs of X and Y are:

P(X=k) = (e^(-α) * α^k) / k!
P(Y=j) = (e^(-β) * β^j) / j!

Now, let's define Z as the random variable that represents the sum of X and Y. We can find the PMF of Z by convolution, which involves summing the joint probabilities of all possible pairs of X and Y that add up to a given value of Z.

P(Z = k) = P(X + Y = k)
= ∑(i=0 to k) P(X = i) * P(Y = k - i)
= ∑(i=0 to k) [(e^(-α) * α^i) / i!] * [(e^(-β) * β^(k-i)) / (k-i)!]
= (e^(-α-β)) / k! * ∑(i=0 to k) (α^i * β^(k-i)) / (i! * (k-i)!)

The last step follows from the fact that the sum in the second line is the product of two Poisson PMFs, which is the PMF of a Poisson random variable with parameter equal to the sum of the two original parameters.

Therefore, we have derived the PMF of Z, which is a Poisson distribution with parameter α+β:

P(Z = k) = (e^(-(α+β))) * ((α+β)^k) / k!

So, the PMF of Z is given by a Poisson distribution with parameter equal to the sum of the parameters of X and Y.
