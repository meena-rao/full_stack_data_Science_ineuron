### 2. What are the different forms of data used in machine learning? Give a specific example for each of them.
Most data can be categorized into 4 basic types from a Machine Learning perspective: Numerical Data, Categorical Data, Time-Series Data, and Text data.
__Numerical Data:__
Numerical data is any data where data points are exact numbers. Statisticians also might call numerical data, quantitative data. This data has meaning as a measurement such as house prices or as a count, such as a number of residential properties in Los Angeles or how many houses sold in the past year.

Numerical data can be characterized by continuous or discrete data. Continuous data can assume any value within a range (ex. 78.87) whereas discrete data has distinct values (ex: 78)

__Categorical Data:__
Categorical data represents characteristics, such as a hockey player’s position, team, hometown. Categorical data can take numerical values. For example, maybe we would use 1 for the colour red and 2 for blue. But these numbers don’t have a mathematical meaning. That is, we can’t add them together or take the average.

__Time Series Data:__
Time series data is a sequence of numbers collected at regular intervals over some period of time. It is very important, especially in particular fields like finance. Time series data has a temporal value attached to it, so this would be something like a date or a timestamp that you can look for trends in time.

__Text:__
Text data is basically just words. A lot of the time the first thing that you do with text is you turn it into numbers using some interesting functions like the bag of words formulation.

### 8. What are the various central inclination measures? Why does mean vary too much from median in certain data sets?
Central tendency measures are statistical measures that describe the central point of a dataset. The three most common measures of central tendency are:

__Mean:__ The mean is the sum of all values in a dataset divided by the number of observations. It is commonly used to represent the average value of a dataset.

_Median:__ The median is the middle value in a sorted dataset. It represents the value that is at the exact center of the dataset when all values are arranged in order.

__Mode:__ The mode is the most frequently occurring value in a dataset. It is commonly used to describe the most common value in a dataset.

When a dataset has a normal distribution, the mean, median, and mode are equal. However, when a dataset is skewed, the mean and median can differ significantly. For example, in a dataset with a positive skew, the mean will be higher than the median because the skewness is pulling the mean towards the larger values. In contrast, in a dataset with a negative skew, the mean will be lower than the median because the skewness is pulling the mean towards the smaller values.

In some datasets, the mean varies too much from the median due to the presence of outliers. Outliers are extreme values that are significantly different from the other values in the dataset. They can have a large impact on the mean, pulling it towards their extreme value. In contrast, the median is not affected by outliers, as it represents the value at the exact center of the dataset. Therefore, when a dataset has outliers, the mean may vary significantly from the median, and the median may be a more appropriate measure of central tendency.

For example, consider a dataset of salaries for employees in a company. If the CEO's salary is included in the dataset, which is much higher than other employees' salaries, the mean will be significantly higher than the median. In this case, the median may be a better measure of central tendency as it represents the value at the exact center of the dataset and is not influenced by the outlier CEO's salary.

### 9. Describe how a scatter plot can be used to investigate bivariate relationships. Is it possible to find outliers using a scatter plot?
A scatter plot is a graph that displays the relationship between two variables in a dataset. It is used to investigate bivariate relationships, which are relationships between two variables. A scatter plot can help to identify patterns, trends, and potential outliers in the data.

To create a scatter plot, the values of the two variables are plotted on the x- and y-axis of the graph. Each point on the graph represents one observation in the dataset. The scatter plot can be used to identify patterns or trends in the relationship between the two variables. For example, if there is a positive relationship between the variables, the points on the scatter plot will tend to form a diagonal line from the lower left to the upper right. If there is a negative relationship between the variables, the points will tend to form a diagonal line from the upper left to the lower right.

A scatter plot can also be used to identify potential outliers in the data. Outliers are data points that are significantly different from other observations in the dataset. They can be identified as points that are located far away from the majority of the points on the scatter plot. Outliers may indicate errors in data collection or may be genuine extreme values. It is important to investigate outliers and determine their cause before deciding how to handle them in the analysis.

Overall, a scatter plot is a useful tool for investigating bivariate relationships and identifying potential outliers in the data. It can provide valuable insights into the relationship between two variables and can help to guide further analysis.

### 10. Describe how cross-tabs can be used to figure out how two variables are related.
Cross-tabulation, also known as contingency table analysis, is a statistical tool used to examine the relationship between two categorical variables. It provides a way of summarizing the data in a table format, showing the frequency and percentage of each combination of categories for the two variables being studied.

To create a cross-tabulation, the categories for one variable are listed in the rows, while the categories for the other variable are listed in the columns. The frequency and percentage of each combination of categories are then computed and entered into the corresponding cell in the table.

Cross-tabulation can be used to answer research questions related to the relationship between two categorical variables. For example, suppose we are interested in examining the relationship between gender and smoking status. We can create a cross-tabulation of gender and smoking status, which will show the frequency and percentage of males and females who smoke or do not smoke. This table can help us to identify any patterns or differences in smoking behavior between males and females.

Cross-tabulation can also be used to identify conditional relationships between variables. For example, we may be interested in examining the relationship between gender and smoking status, while also taking into account age as a third variable. We can create a cross-tabulation of gender and smoking status for each age group, which will help us to determine if the relationship between gender and smoking status is different for different age groups.

Overall, cross-tabulation is a powerful tool for analyzing relationships between two categorical variables. It can provide valuable insights into the relationships between variables and can help researchers to identify important patterns and differences in the data
