```python
class bag_of_words:        
    def word_repeat(self, file):
        '''Takes file name as input and returns count of how many times each word occurs'''
        try:
            d = dict()
            with open(file,'r') as f:
                line = f.read()
                words = line.split()
                
                for word in words:
                    if word in d:
                        d[word] = d[word] + 1
                    else:
                        d[word] = 1
                        
            #print(f'Details of words occorance in {f}:\n')
            #print(d.items())
            logging.info(f'Details of how many times each word repeats in {f}:\n {d.items()}')
            return d.items()
        except Exception as E:
            logging.error("\nerror occured")
            logging.exception(str(e))
        finally:
            f.close()                        
            
            
    def word_start_with_same_alpha(self, file):
        '''Takes file name as input and returns count of words start with same alphabet'''
        try:
            d = dict()
            with open(file,'r') as f:
                line = f.read()
                words = line.split()
                
                for word in words:
                    if word[0] in d:
                        d[word[0]] = d[word[0]] + 1
                    else:
                        d[word[0]] = 1
                        
            logging.info(f'Details of ount of words start with same alphabet in file {f}: \n {sorted(d.items())}')
            return sorted(d.items())
        
        except Exception as E:
            logging.error("\nerror occured")
            logging.exception(str(e))

        finally:
            f.close()
            
            
            
    def filter_f(self,file):
        '''takes filename as i/p and lists words starting with only alphabets'''
        import re
        try:
            f2 = open('f_new.txt', 'w+',encoding = 'utf-8')
            with open(file, 'r', encoding = 'utf-8') as f1:
                data = f1.read()
                tokens = data.strip().split()
                for word in tokens:
                    n_word = re.sub("[^a-zA-Z]","",word)
                    if n_word != '':
                        f2.write(n_word)
                        f2.write("\n")
                f2.seek(0)
                data = f2.read()
                #print(data)
                logging.info(f'List of words only with alphabets without special characters:{data}')
                return data
        except Exception as E:
            logging.error("went wrong")
            logging.exception(str(e))

        finally:
            try:
                f1.close()
                f2.close() 
            except Exception as E:
                logging.error("error occured")
                logging.exception(str(e))

            
    def tuple_from_all_files(self, f1,f2,f3,f4,f5):
        '''create tuple of all the records available in all the files'''
        try:                    
            def file_to_list(file):
                '''Takes a file as i/p and returns list of words in it'''
                try:
                    li= list()
                    fp =open(file,'r', encoding='utf-8')
                    data = fp.read()
                    tokens = data.strip().split()
                    for token in tokens:
                        li.append(token)                    
                    return li
                except Exception as E:
                    logging.error("error occured")
                    logging.exception(str(e))
                finally:
                    fp.close()
        
            l1 = file_to_list(f1)
            l2 = file_to_list(f2)
            l3 = file_to_list(f3)
            l4 = file_to_list(f4)
            l5 = file_to_list(f5)
            logging.info('All Files convered to lists successfully')
            logging.info(f'Tuple of all the records available in all the files {list(zip(l1,l2,l3,l4,l5))}')
            return(list(zip(l1,l2,l3,l4,l5)))

        except Exception as E:
            logging.error("Error occured :")        
            logging.exception(str(e))

       
    def list_to_sqlite(self, lis):
        '''Takes list of tuples as input and inserts into SQLite3 database'''
        try:
            import sqlite3
            conn = sqlite3.connect("challenge1.db")
            cur = conn.cursor()
        
            cur.execute('CREATE TABLE words_table2(w1 TEXT, w2 TEXT, w3 TEXT, w4 TEXT, w5 TEXT)')
            cur.executemany('INSERT INTO words_table2 VALUES (?, ?, ?, ?, ?)', lis)
            conn.commit()
        
            data = cur.execute("SELECT * FROM words_table2 LIMIT 5")
            for i in data:
                #print(i)
                logging.info(i)
                
        except Exception as E:
            logging.error("error occured")
            logging.exception(str(e))
        finally:
            cur.execute('DROP TABLE IF EXISTS words_table2')
            cur.close()
            conn.close()
```
